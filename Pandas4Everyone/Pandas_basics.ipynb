### Using Group By

```python
fig = plt.figure()

# Divide the figure into a 1x2 grid, and give me the first section
ax1 = fig.add_subplot(121)

# Divide the figure into a 1x2 grid, and give me the second section
ax2 = fig.add_subplot(122)

df.groupby('sex')['tip'].mean().plot(kind = 'pie',  ax= ax1)
df.groupby('time')['tip'].mean().plot(kind = 'pie',  ax= ax2)

plt.title('Tip Proportions by Sex and time', size = 15)
None
```

Males tip more and dinner time account for more tips

```python
# Getting all values where tips > mean populaton tip
above_average = df[df.tip > df.tip.mean()]
above_average.shape
```
