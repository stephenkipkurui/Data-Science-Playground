### Why vizuals? 
English Statitician Frank Anscombe data show set with four continuous values each having same mean, variance, correlation, and regression line. However on plotting, its evident that these same values are not really same.

```python
# Get libraries
from libraries import*
```

```python
# Get Data
anscombe = sns.load_dataset('anscombe')
anscombe
```

```python
# Get subsets
set_I = anscombe[anscombe.dataset == 'I']
set_II = anscombe[anscombe.dataset == 'II']
set_III = anscombe[anscombe.dataset == 'III']
set_IV = anscombe[anscombe.dataset == 'IV']
```

```python
# Mean for all sets

print(f'Set I\n{set_I.mean()}\n\n Set II\n\
{set_II.mean()}\n\n Set III\n {set_III.mean()}\
\n\n Set IV\n{set_IV.mean()}')
```

```python
# Create main figure to hold all subplots
fig = plt.figure()

ax1 = fig.add_subplot(2,2, 1)
ax2 = fig.add_subplot(2,2, 2)
ax3 = fig.add_subplot(2,2, 3)
ax4 = fig.add_subplot(2,2, 4)

#Plot
ax1.plot(set_I.x, set_I.y)
ax2.plot(set_II.x, set_II.y)
ax3.plot(set_III.x, set_III.y)
ax4.plot(set_IV.x, set_IV.y)

# Add titles
ax1.set_title('Set I')
ax2.set_title('Set II')
ax3.set_title('Set III')
ax4.set_title('Set IV')

plt.show()
```
