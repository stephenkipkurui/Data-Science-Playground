### Why visuals? 
English Statitician Frank Anscombe data show set with four continuous values each having same mean, variance, correlation, and regression line. However on plotting, its evident that these same values are not really same.

```python
# Get libraries
from libraries import*
```

```python
# Get Data
anscombe = sns.load_dataset('anscombe')
anscombe
```

```python
# Get subsets
set_I = anscombe[anscombe.dataset == 'I']
set_II = anscombe[anscombe.dataset == 'II']
set_III = anscombe[anscombe.dataset == 'III']
set_IV = anscombe[anscombe.dataset == 'IV']
```

```python
# Mean for all sets

print(f'Set I\n{set_I.mean()}\n\n Set II\n\
{set_II.mean()}\n\n Set III\n {set_III.mean()}\
\n\n Set IV\n{set_IV.mean()}')
```

```python
# Create main figure to hold all subplots
fig = plt.figure()

ax1 = fig.add_subplot(2,2, 1)
ax2 = fig.add_subplot(2,2, 2)
ax3 = fig.add_subplot(2,2, 3)
ax4 = fig.add_subplot(2,2, 4)

#Plot
ax1.plot(set_I.x, set_I.y)
ax2.plot(set_II.x, set_II.y)
ax3.plot(set_III.x, set_III.y)
ax4.plot(set_IV.x, set_IV.y)

# Add titles
ax1.set_title('Set I')
ax2.set_title('Set II')
ax3.set_title('Set III')
ax4.set_title('Set IV')

# fig.tight_layout()

plt.show()
```

## (i). Matplotlib

### Univariate 

Use histograms

```python
tips.hist()
None
```

### Bi-variate

- For Continuous variables, use scatterplot
- For discrete and continuous variables use boxplots

```python
# Continuous vars
plt.scatter(tips.total_bill, tips.tip)
plt.title('Total Bill vs Tips Amount')
plt.xlabel('Total Bill')
plt.ylabel('Tips')
None
```

```python
# Discrete vars
male_tips = tips[tips['sex'] == 'Female']['tip'], 
female_tips = tips[tips['sex'] == 'Male']['tip'], 
labels =  ['Female', 'Male']
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1,1,1)
axes1.boxplot(male_tips, female_tips)
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Tips by Sex')
boxplot.show()
```

### Multi-variate

```python
def impute_gender(sex):
    
    if sex == 'Female':
        return 0
    else:
        return 1
    
tips['sex_color'] = tips['sex'].apply(impute_gender)

scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x = tips['total_bill'],
    y = tips['tip'],
    c = tips['sex_color'],
    alpha = 0.5)
axes1.set_title('Total Bill vs Tips by Sex')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
scatter_plot.show()
```

## (ii). Seaborn 

### Univariate
In seaborn, histograms are created using `sns.distplot`.

Default distplot plots both histogram and density plot (line in graph below). Setting `kde` to False will create histogram.

KDE = Kernel Density Estimation

#### Default Plot

```python
## Univariate 
hist, ax = plt.subplots()
ax = sns.distplot(tips['total_bill'])
ax.set_title('Total Bill Frequency')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
plt.show()
```

#### Histogram

```python
# Plotting just histogram. Setting kde = False

hist, ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], kde = False)
ax.set_title('Total Bill Frequency')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
plt.show()
```

### kde

```python
# Plotting just kde. Setting hist = False
#
# area under curve is 1. Distribution centered at each data point
#

hist, ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], hist = False)
ax.set_title('Total Bill Frequency')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
plt.show()
```

#### Rug plot (in addition to kde & histogram)

Used with other plot to enhance visuals

```python

hist, ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], rug = True)
ax.set_title('Total Bill Frequency')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
plt.show()
```

#### Countplot

Counts discrete variables

```python

hist, ax = plt.subplots()
ax = sns.countplot(tips['day'])
ax.set_title('Total Bill Frequency')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
plt.show()
```

### Bi-variate 
In seaborn, regplot is used to create bivariate plots

setting `set_reg = False` removes regression line

#### Reg-plot

```python
hist, ax = plt.subplots()
ax = sns.regplot(x = 'total_bill', y = 'tip', data =  tips)
ax.set_title('Total Bill Frequency')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
plt.show()
```

#### lm-plot

Internally lmplot calls regplot. Difference is regplot creates axes wheres lmplot creates fugures.

```python
fig = sns.lmplot(x = 'total_bill', y = 'tip', data = tips)

plt.show()
```

#### Joint Plot

```python
fig = sns.lmplot(x = 'total_bill', y = 'tip', data = tips)

plt.show()
```

#### Hexbin Plot

A more cleaner visual for presenting binned variables- 2D representation. 

```python
hexbin = sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'hex')
hexbin.set_axis_labels(xlabel = 'Total Bill', ylabel = 'Tip')
plt.show()
```

#### Kde 

```python
hexbin = sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'kde')
hexbin.set_axis_labels(xlabel = 'Total Bill', ylabel = 'Tip')
plt.show()
```

#### Pair Plot Graphs


Plots a relationship for numeric variables. Pairplot alone creates redundant information for 1/2 of the chart.


```python
fig = sns.pairplot(tips)
```

- Using `pairgrid`  to manualy assign plots section 

```python
pair_grid = sns.PairGrid(tips)
pair_grid = pair_grid.map_upper(sns.regplot)
pair_grid = pair_grid.map_lower(sns.kdeplot)
pair_grid = pair_grid.map_diag(sns.distplot, rug = True)
```
